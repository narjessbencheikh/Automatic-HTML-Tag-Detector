**The word cloud will display the most frequent words in the text, with a larger font size
for the most frequent words.**

```python
import pandas as pd
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt


# Get a list of all unique labels/categories
labels = df['label'].unique()

# Set the number of columns and rows for the subplots
num_cols = 3
num_rows = int(len(labels) / num_cols) + (1 if len(labels) % num_cols != 0 else 0)

# Create the subplots
fig, axs = plt.subplots(num_rows, num_cols, figsize=(20, 10))

# Generate a word cloud for each label/category and add it to the corresponding subplot
for i, label in enumerate(labels):

    data = df[df['label'] == label]

    # Extract the list of words associated with the current label/category
    words = list(data['words'])

    # Count the frequency of each word
    word_counts = Counter(words)

    # Generate the word cloud for the current label/category
    wordcloud = WordCloud(background_color='white')
    wordcloud.generate_from_frequencies(word_counts)

    # Add the word cloud to the corresponding subplot
    row_idx = i // num_cols
    col_idx = i % num_cols
    axs[row_idx, col_idx].imshow(wordcloud, interpolation='bilinear')
    axs[row_idx, col_idx].axis("off")
    axs[row_idx, col_idx].set_title(label, fontsize=20)


fig.subplots_adjust(hspace=0.5, wspace=0.5)


plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Création d'un dictionnaire pour stocker les mots pour chaque tag
tag_word_dict = {}
for i, row in df.iterrows():
    tag = row['tag']
    words = row['words'].split()
    if tag in tag_word_dict:
        tag_word_dict[tag] += words
    else:
        tag_word_dict[tag] = words

# Création d'un graphique en barres pour représenter la distribution des mots pour chaque tag
plt.figure(figsize=(15, 5))
for tag, words in tag_word_dict.items():
    word_count = pd.Series(words).value_counts().sort_values(ascending=False)[:10]
    plt.bar(word_count.index, word_count.values, alpha=0.5, label=tag)

plt.legend()
plt.title('Words Distribution Across Tags')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.show()
```

```python
import nltk
from nltk.corpus import stopwords

# download stopwords
nltk.download('stopwords')

# create a set of stopwords
stop_words = set(stopwords.words('english'))

# create a dictionary to store columns containing stop words for each label
stop_words_dict = {}

for label in df['label'].unique():
    # create a set to store the stop words for this label
    label_stop_words = set()
    for i, row in df[df['label']==label].iterrows():
        # tokenize the words in the row
        words = row['words'].split()
        # check if any of the words in the row are in the stop words set
        for word in words:
            if word in stop_words:
                label_stop_words.add(word)
    # add the set of stop words for this label to the dictionary
    stop_words_dict[label] = label_stop_words


for label in stop_words_dict:
    print('Stop words for label {}:'.format(label))
    print(stop_words_dict[label])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from string import punctuation
from collections import Counter

# Extract the punctuation marks
punctuations = set(punctuation)

# Filter the words that are punctuation marks
punctuations_count = Counter([char for word in df["words"]
                              for char in word
                              if char in punctuations])

punctuations_count
```

```python
# Create a bar plot of the punctuation marks frequency
plt.figure(figsize=(10, 5))
plt.bar(punctuations_count.keys(), punctuations_count.values())
plt.xticks(rotation=45, ha="right")
plt.title("Punctuation Marks Frequency")
plt.xlabel("Punctuation Mark")
plt.ylabel("Frequency")
plt.show()
```

```python
import pandas as pd
import string
import matplotlib.pyplot as plt

def count_punctuation(text):
    count = 0
    for char in text:
        if char in string.punctuation:
            count += 1
    return count


# Loop over each label and count the punctuation
for label in df["label"].unique():
    # Filter the dataset for the current label
    label_df = df[df["label"] == label]

    # Calculate the punctuation count for each word in the "words" column
    punctuation_count = label_df["words"].apply(count_punctuation)

    # Create a histogram of the punctuation count
    plt.hist(punctuation_count, bins=20, alpha=0.5, label=label)

# Set the plot title and labels
plt.title("Punctuation Distribution by Label")
plt.xlabel("Punctuation Count")
plt.ylabel("Frequency")

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

## **Stemming OR Lemmatization ?**

For the NER task, lemmatization is generally considered to be better than stemming because it produces the base form of the word, which can help maintain the context and meaning of the text. Stemming, on the other hand, only removes the suffixes of the words, which can sometimes result in words with different meanings being reduced to the same root form.

For example, the words "running" and "ran" would both be reduced to "run" by a stemming algorithm, while lemmatization would produce "run" and "ran" respectively. In the context of NER, maintaining the correct form of the word can be important for identifying named entities correctly.
**==> Lemmatization**

```python
def assign_pos_tags(sentence):
    tokens = word_tokenize(sentence)
    pos_tags = pos_tag(tokens)
    return [tag for _, tag in pos_tags]
```
